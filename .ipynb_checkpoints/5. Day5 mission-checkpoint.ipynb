## 2-1. 비트코인 가격 그래프

```python
bc = pd.read_csv('./dataset/BitCoin.csv')

bc
```

```python
# 타입 확인
bc.dtypes
```

```python
# 날짜 순으로 정렬
bc = bc.sort_values(by=['Date'], ignore_index=True)

bc
```

```python
# 기간 확인
bc[(bc['Date'] < '2017-06') & ('2016-06' < bc['Date'])]
```

5일 평균선 그리기

```python
col = bc['Open']
added = pd.Series([0]*len(col))

for i in range(1, 5):
    added += pd.Series([0]*i).append(col[:-i], ignore_index=True)

print(col + added)
```

```python
ma5 = col + added

for i in range(5):
    ma5[i] = ma5[i]/(i+1)
for i in range(5, len(col)):
    ma5[i] = ma5[i]/5

ma5
```

```python
bc['ma5'] = ma5

bc.columns
```

```python
# 기간 픽
bc_term = bc[(bc['Date'] < '2017-06') & ('2016-06' < bc['Date'])]

bc_term
```

그래프 그리기

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = range(len(bc_term['ma5']))
y = bc_term['ma5']

plt.figure(figsize=(12, 6))

plt.plot(x, y, color='#f2a900')

plt.xlabel('Date')
plt.ylabel('ma5')

plt.show()
```
