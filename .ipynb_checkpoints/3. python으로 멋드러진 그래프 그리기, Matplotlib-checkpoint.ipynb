## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- %matplotlib inline을 통해서 활성화

cf) 라이브러리 vs 프레임워크

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### 2. Case Study with Arguments

```python
plt.plot([1, 2, 3, 4, 5]) # y = x + 1

plt.show()
```

### Figsize: Figure(도면)의 크기 조절

```python
plt.figure(figsize=(6, 6)) # 1당 72pixel

plt.plot([1, 2, 3, 4, 5]) # y = x + 1
plt.show()
```

### 2차함수 그래프 with plot()

```python
# 리스트를 이용해서 1차함수 y = x 그리기
plt.plot([0, 1, 2, 3, 4])
plt.show()
```

```python
# np.array를 이용해서 함수 그리기

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])

plt.plot(x, y)
plt.show()
```

```python
# np.arange(a, b, c)

x = np.arange(-10, 10, 0.01)

plt.plot(x, x**2)
plt.show()
```

```python
# x. y축에 설명 추가하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축 범위 설정

x = np.arange(-10, 10, 0.01)

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.axis([-5, 5, 0, 20])

plt.plot(x, x**2)
plt.show()
```

```python
# x. y축에 눈금 설정

x = np.arange(-10, 10, 0.01)

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.axis([-5, 5, 0, 25])

plt.xticks([i for i in range(-5, 6)])
plt.yticks([i for i in range(0, 27, 3)])

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프에 title 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.axis([-5, 5, 0, 25])

plt.xticks([i for i in range(-5, 6)])
plt.yticks([i for i in range(0, 27, 3)])

plt.title("y = x^2 graph")

plt.plot(x, x**2, label="trend")
plt.legend()

plt.show()
```

## II. Matplotlib Case Study

## 3. Matplotlib Case Study

### 꺾은선 그래프 (Plot)

```python
x = np.arange(20)
y = np.random.randint(0, 20, 20)

plt.plot(x, y)

plt.axis([0, 20, 0, 20])
plt.yticks([0, 5, 10, 15, 20])

plt.show()
```

### 산점도 (Scatter Plot)

- `.scatter()`

```python
plt.scatter(x, y)
plt.show()
```

시계열 vs 상관관계

### 박스 그래프 (Box Plot)

- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)

```python
plt.boxplot(y)

plt.show()
```

```python
plt.boxplot((x, y))

# title
plt.title("Box plot of x, y")

plt.show()
```

### 막대 그래프 (Bar Plot)
- 범주형 데이터의 값과 그 크기를 직사각형으로 나타냄
- `.bar()`

```python
plt.bar(x, y)

plt.xticks(np.arange(0, 21, 1))

plt.show()
```

```python
# cf) Histogram
# .hist()
# 도수분포를 직사각형의 막대 형태로 나타냈다.
# "계급"으로 나타낸 것이 특징: 0,1,2 -> 0~2

plt.hist(y, bins=np.arange(0, 21, 2))
plt.xticks(np.arange(0, 21, 2))

plt.show()
```

### 원형 그래프 (Pie Chart)

- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타냄
- 다른 그래프에 비해서 **비율** 확인에 용이
- `.pie()`

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['1', '2', '3', '4'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### Seaborn import 하기

```python
import seaborn as sns
```

### 커널밀도그림 (Kerner Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- `sns.kdeplot()`

```python
# Histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins=x)

plt.show()
```

```python
# kdeplot

sns.kdeplot(y, shade=True)

plt.show()
```

### 카운트그림 (Count Plot)

- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({'name':["Andy", "Bob", "Cat"], "vote":[True, True, False]})

vote_df
```

```python
# matplot

vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])

plt.show()
```

```python
# sns countplot

sns.countplot(x=vote_df['vote'])

plt.show()
```

### 캣그림 (Cat Plot)

- 숫자형 변수화 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x="WHO Region", y="Confirmed", data=covid)
# hue, kind keyword also exist

s.fig.set_size_inches(10, 6)
plt.show()
```

### 스트립그림 (Strip Plot)

- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`

```python
sns.stripplot(x="WHO Region",y="Recovered",data=covid)

plt.show()
```

```python
# cf) swarmplot

sns.swarmplot(x="WHO Region",y="Recovered",data=covid)

plt.show()
```

### 히트맵 (Heatmap)

- 데이터의 행렬을 색상으로 표현해주는 그래프
- `sns.heatmap()`

```python
covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```
